```python
def get_track_ids(playlist_id):
    music_id_list = []
    playlist = sp.playlist(playlist_id)
    
    for item in playlist['tracks']['items']:
        music_track = item['track']
        music_id_list.append(music_track['id'])
    return music_id_list    
```

```python
def get_track_data(track_id):
    meta = sp.track(track_id)
    track_details = {'name': meta['name'], 'album': meta['album']['name'],'artist': meta['album']['artists'][0]['name'], 'release_date': meta['album']['release_date'], 'duration_in_mins': round((meta['duration_ms']* 0.001) / 60.0,2) 
    }
    return track_details
```

```python
#playlist_id = input('Enter the playlist id')
track_ids = get_track_ids(playlist_id)
print(len(track_ids))
print(track_ids)

tracks = []
for i in range(len(track_ids)):
    time.sleep(5)
    track = get_track_data(track_ids[i])
    tracks.append(track)
```

```python
with open('spotify_playlist.json', 'w') as outfile:
    json.dump(tracks, outfile, indent = 4)
```
